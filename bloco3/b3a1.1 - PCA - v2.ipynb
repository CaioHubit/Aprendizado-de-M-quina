No entanto, em casos reais é muito difícil encontrarmos poucas componentes principais que representem bem os dados, então como sabemos quantas componentes utilizarmos? Para isso podemos utilizar a variância explicada, ela medirá, a grosso modo, o quanto dos seus dados são explicados com cada componente principal. 

Matemáticamente, a soma das variâncias da matriz de co-variância original é a mesma que a da nova matriz obtida após a aplicação da transformação linear, se levarmos em conta todas as componentes principais. Mas assim não estamos reduzindo a dimensionalidade do problema. Como a transformação linear aplicada, leva a algo parecido com uma linearização (confiram com calma o vídeo do StatQuest), isso leva a uma maior variância para a primeira PC1, depois para a segunda e assim por diante. Logo, ao graficar a variância de cada componente principal de um dataset qualquer, teremos o seguinte:

<img src="https://miro.medium.com/proxy/0*_m1sJvWw4JZ3qMB-.png" width="400"/>

Notem que a primeira componente principal (PC1), tem alta variância, as outras cada vez menores. Depois de um certo ponto há pouco ganho em utilizar mais componentes. Assim, utilizando 6 das componentes principais, já explicamos 90% dos dados. Por isso esse tipo de gráfico é chamado de variância explicada. ## Voltando ao dataset dos diamantes

Vamos voltar ao nosso exemplo do dataset de diamantes. Vamos carregar o dataser como fizemos nas outras aulas:

```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

DATASET_NAME = "diamonds"
df = sns.load_dataset(DATASET_NAME)
FEATURES = ["carat", "depth", "table", "x", "y", "z"]

X = df.reindex(FEATURES, axis=1)
```

Como já dito, é importante normalizar os dados, para que uma componente de dimensão maior não domine sua análise PCA. Façamos isso então:

```python
sc = StandardScaler()
sc.fit(X)
X_norm = sc.transform(X)
```

Agora vamos rodar a PCA em si, agora não estamos utilizando um regressor ou classificador, mas estamos realizando uma transformação. Abaixo segue o cálculo da PCA:

```python
pca = PCA()
pca.fit(X_norm)
X_pca = pca.transform(X_norm)
```

Vamos graficar a PCA obtida, junto dos nossos dois primeiros eixos de features, para entendermos o que a PCA fez:

```python
#X_new = pca.inverse_transform(X_pca)
plt.scatter(X_norm[:, 0], X_norm[:, 1], alpha=0.2)
#plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.1)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.axis('equal');
```

Podemos notar que a PCA "espalhou" um pouco mais nossos dados, no fim isso aumenta a variância do primeiro eixo no novo sistema de coordenadas.

Como sabemos os preços dos nossos diamantes, neste caso podemos colorir os pontos de acordo com o preço, para entender se a PCA agrupou os dados de alguma maneira.

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], edgecolor='none', alpha=0.5, c=df["price"].values)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
```

É possível notarmos uma certa ordem, com os maiores valores a direita e os menores valores a esquerda, porém alguns pontos ficam fora da tendência ao utilizarmos somente as duas primeiras componentes principais.

Vamos fazer um gráfico com as três componentes principais então:

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=df["price"].values)
```

Não parece que melhoramos muito a representação nesse caso, pois os pontos que estavam fora da tendência ainda continuam fora da tendência. Isso indica que incluir uma terceira componente principal não está trazendo tantos ganhos.

Vamos agora investigar o que cada componente principal levou em conta em relação as *features* originais. Para isso vamos averiguar a matriz de *loadings*:

```python
component_names = [f"PC{i+1}" for i in range(X_pca.shape[1])]
X_pca = pd.DataFrame(X_pca, columns=component_names)
X_pca.head()

loadings = pd.DataFrame(
    pca.components_.T,  # transpose the matrix of loadings
    columns=component_names,  # so the columns are the principal components
    index=X.columns,  # and the rows are the original features
)
loadings
```

Podemos notar que a PC1 teve um grande peso para $carat$, $x$, $y$ e $z$. Um pouco para $table$, e praticamente nada para $depth$. Já a PC2 teve seus maiores pesos para $depth$ e $table$. Já a PC3 também, porém praticamente invertendo o peso entre $table$ e $depth$.

Ok, agora vamos graficar a variância explicada, como dito na primeira parte da aula, ela ajuda a entender o quanto de sua amostra está sendo representada com cada componente principal.

```python
def plot_variance(pca, width=8, dpi=100):
    # Criar figura
    fig, axs = plt.subplots(1, 2)
    n = pca.n_components_
    grid = np.arange(1, n + 1)
    
    # Fazer o cálculo da variância explicada
    evr = pca.explained_variance_ratio_
    axs[0].bar(grid, evr)
    axs[0].set(
        xlabel="Componente", 
        title="Variancia explicada", 
        ylim=(0.0, 1.0)
    )
    
    # Variância acumulada
    cv = np.cumsum(evr)
    axs[1].plot(np.r_[0, grid], np.r_[0, cv], "o-")
    axs[1].set(
        xlabel="Componente", 
        title="Variância acumulada", 
        ylim=(0.0, 1.0)
    )
    
    # Detalhes da figura
    fig.set(figwidth=8, dpi=100)
    return axs

# Graficar
plot_variance(pca)
```

A variância explicada nos mostra que a PC1 é a principal para capturar os dados, já com a PC2 e a PC3 estamos explicando praticamente todos os dados. Ou seja, ao invés de utilizarmos seis *features*, poderemos utilizar somente três, simplificando os modelos.
